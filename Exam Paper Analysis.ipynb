Introduction The midmarks dataset provides a detailed record of students' midterm exam scores across various subjects. This dataset is useful for analyzing student performance, identifying trends in different sections, and evaluating subject-wise difficulty levels. By studying the data, we can assess which subjects students excel in and which require further attention. Additionally, analyzing variations across sections can help educators refine their teaching methods and provide better academic support.

```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_excel("MIDMARKS.xlsx")
df.fillna(0, inplace=True)
df["S.NO"]=df["S.NO"].astype("int64")
df["S.NO"]=range(1,len(df)+1)
df=df.replace("MP",0)
df=df.replace("A",0)
df=df.replace("AB",0)
df
```

```python
df["Total"]=df["DV"]+df["M-II"]+df["PP"]+df["BEEE"]+df["FL"]+df["FIMS"]

def assign_grade(total):
    if total >= 110:
        return 'A+'
    elif total >= 100:
        return 'A'
    elif total >= 90:
        return 'B'
    elif total >= 80:
        return 'C'
    elif total >= 70:
        return 'D'
    else:
        return 'F'
df['Grade'] = df['Total'].apply(assign_grade)
df
```

```python
df['SECTION'] = df['SECTION'].fillna('SIGMA')  
df['SECTION'] = df['SECTION'].replace('', 'SIGMA')
df['SECTION'] = df['SECTION'].replace(0, 'SIGMA') 
df
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

top_10 = df[df["Total"] >= df["Total"].quantile(0.9)]
bottom_10 = df[df["Total"] <= df["Total"].quantile(0.1)]

top_bottom_means = pd.concat([
    top_10[["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]].mean(),
    bottom_10[["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]].mean()
], axis=1)
top_bottom_means.columns = ["Top 10%", "Bottom 10%"]

top_bottom_means.plot(kind="bar", figsize=(10, 6), colormap="coolwarm", edgecolor="pink")
plt.title("Subject Scores: Top 10% vs Bottom 10% Performers")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.legend()
plt.show()
```

The code calculates the average scores for the top 10% and bottom 10% of students in each subject, then visualizes the comparison as a bar chart with the "coolwarm" colormap and pink edges.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
section_averages = df.groupby("SECTION")[["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]].mean()

plt.figure(figsize=(10, 6))
sns.heatmap(section_averages, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Average Scores by SECTION and Subject")
plt.show()

lowest_performers = section_averages.mean(axis=1).idxmin()
print(f"The section that underperforms across all subjects is: {lowest_performers}")

```

The code calculates the average scores by section for each subject, visualizes them as a heatmap, and then identifies and prints the section with the lowest overall performance across all subjects.
